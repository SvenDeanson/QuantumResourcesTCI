# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.cm as cm
from matplotlib import rc

# Set LaTeX font for the plot
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(11, 7))

# Set labels with LaTeX
ax.set_xlabel(r'$h$', fontsize=40)
ax.set_ylabel(r'$\mathcal{C}$', fontsize=40)

# Set tick label size
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set_xticks([0, 0.5, 1, 1.5, 2.0])

# Define parameters
J = -1.0
N_list = [(3, 3), (4, 4), (5, 5),(6, 6), (7, 7), (8, 8)]

markerstyles = ['.','o','s','D','^',"X"]

default_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
brown_hex = "#8c564b"
filtered_colors = [color for color in default_colors if color.lower() != brown_hex.lower()]


for idx,(Nx, Ny) in enumerate(N_list):
    path = f"./Ising_2D_Data/Nx_{Nx}_Ny_{Ny}/"
    output_csv = path + f"M_Ising_2D_Nx_{Nx}_Ny_{Ny}_J_{J}.csv"
    
    data = pd.read_csv(output_csv)
    
    h_values = data['h']
    coherence_results = data['Coherence']
    
    ax.plot(h_values, coherence_results, marker=markerstyles[idx], linestyle='-', alpha=0.75, linewidth=3, markersize=8,
            label=rf'$L={Nx} \times {Ny}$',color=filtered_colors[idx])

ax.legend(frameon=False, loc='upper left', fontsize=20)
ax.set_ylim(-1,30)
ax.axhline(0,color="black",alpha=0.5)
plt.show()
```
