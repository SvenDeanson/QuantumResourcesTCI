# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rc
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Set LaTeX font for the plot
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
```

```python
fig, ax = plt.subplots(figsize=(11, 7))

J = -1.0
N_list = [4, 8, 16, 32, 64]
h_values = np.arange(0.01, 2.11, 0.1)

markerstyles = ['.','o','s','D','^']


for idx,N in enumerate(N_list):
    path = f"./Ising_1D_Data/N_{N}/"
    output_csv = path + f"M_Ising_N_{N}_J_{J}.csv"

    data = pd.read_csv(output_csv)
    h_values_csv = data['h']
    magic_results = data['Magic']

    ax.plot(h_values_csv, magic_results, label=rf'$L={N}$', marker=markerstyles[idx], linestyle='-', alpha=0.75, linewidth=3, markersize=8)


ax.set_xticks([0, 0.5, 1, 1.5, 2.0])
ax.axhline(0, color="black", alpha=0.3)
ax.tick_params(axis='both', which='major', labelsize=20)

ax.set_xlabel(r'$h$', fontsize=40)
ax.set_ylabel(r'$\mathcal{M}_{2}$', fontsize=40)
ax.legend(loc='upper left', fontsize=20,frameon=False)

plt.show()
```
